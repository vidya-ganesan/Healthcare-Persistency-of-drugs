```python
drug_df=drug_df.drop_duplicates()
```

```python
drug_df=drug_df.dropna()
```

```python
#Changing column names to lowercase
drug_df.columns=[x.lower().strip() for x in drug_df.columns]
drug_df.columns
```

```python
drug_df.drop(['ptid'], axis=1,inplace=True)
```

```python
#numeric columns present in dataset
nums_col=drug_df._get_numeric_data().columns
nums_col
print('Number of numerical variables:',len(nums_col))
```

```python
#categorical columns 
cat_features=[feature for feature in drug_df.columns if drug_df[feature].dtypes=='O' ]
cat_features
```

```python
for col in cat_features:
    print(drug_df[col].value_counts())
```

```python
for col in cat_features:
    sns.set_theme(style='whitegrid')
    sns.countplot(x=col,data=drug_df)
    plt.title('Countplot of '+ col)
    plt.ylabel('Count')
    plt.show()
```

```python
#Removing Outliers
for col in nums_col:
    print(col)
    Q1=drug_df[col].quantile(0.25)
    Q3=drug_df[col].quantile(0.75)
    IQR=Q3-Q1
    print(Q1)
    print(Q3)
    print(IQR)
    Lower_Whisker=Q1-1.5*IQR
    Upper_Whisker=Q3+1.5*IQR
    print(Lower_Whisker,Upper_Whisker)
    drug_df=drug_df[drug_df[col]<Upper_Whisker]
```

```python
#boxplot with outliers removed
for col in nums_col:
        sns.set_style('whitegrid')
        sns.boxplot(x='persistency_flag',y=col,data=drug_df)
        plt.title('Boxplot of '+col)
        plt.ylabel(col)
        plt.show()
        
```
